```python
import json
import requests

json_data = requests.get("https://jsonplaceholder.typicode.com/todos").json()
json_data
```

```python
from langchain_text_splitters import RecursiveJsonSplitter
json_splitter = RecursiveJsonSplitter(max_chunk_size=300)
json_chunks = json_splitter.split_json({"data": json_data})
```

```python
docs = json_splitter.create_documents(texts=json_chunks)
for doc in docs[:3]:
    print(doc)
```
